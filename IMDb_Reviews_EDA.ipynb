## __Workflow__
### __1. Load and Inspect Dataset__
- Load CSV using pandas
- Inspect the structure, data types, and sample records


```python
import pandas as pd
df = pd.read_csv(r"C:\Users\BDO NARKANDA\Downloads\IMDB Dataset.csv.zip")

# Inspect dataset
print(df.head())
print(df.info())
print(df.shape)
```

### __2. Missing Values Check__
- Identify and handle any missing or null values to ensure dataset integrity

```python
print(df.isnull().sum())
```

### __3. Text Statistics__
- Compute review lengths (number of characters)
- Summarize key metrics such as mean, min, and max review lengths

```python
# Add column for review length
df['review_length'] = df['review'].apply(len)

# Summary statistics
print(df['review_length'].describe())
```

### __4. Data Visualization__
- Histogram of review lengths to understand distribution
- Bar chart for sentiment distribution (positive vs negative)

```python
import matplotlib.pyplot as plt

plt.hist(df['review_length'], bins=50, color='skyblue')
plt.xlabel('Review Length')
plt.ylabel('Frequency')
plt.title('Distribution of Review Lengths')
plt.show()
```

### __5. Exploratory Text Analysis__
- Keyword frequency analysis
- Compare average review lengths by sentiment
- Optional: word clouds for positive vs negative reviews

```python
import seaborn as sns

sns.countplot(x='sentiment', data=df)
plt.title('Sentiment Distribution')
plt.show()
```
